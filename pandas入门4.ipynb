```python
#算数和数据对齐
```

```python
s1 = pd.Series([7.3, -2.5, 3.4, 1.5], index=['a', 'c', 'd', 'e'])
```

```python
s2 = pd.Series([-2.1, 3.6, -1.5, 4, 3.1], index=['a', 'c', 'e', 'f', 'g'])
```

```python
s1
```

```python
s2
```

```python
s1 + s2      #如果存在某个索引对不相同，则返回结果的索引将是索引对的并集。
             #没有交叠的标签位置上，内部数据对齐会产生缺失值。
```

```python
df1 = pd.DataFrame(np.arange(9.).reshape((3, 3)), columns=list('bcd'), 
                   index=['Ohio','Texas', 'Colorado'])
```

```python
df1
```

```python
df2                #行和列都会执行对齐
```

```python
df1 + df2
```

```python
#使用填充值的算数方法
```

```python
df1 = pd.DataFrame(np.arange(12.).reshape((3, 4)), columns = list('abcd'))
```

```python
df2 = pd.DataFrame(np.arange(20.).reshape((4, 5)), columns = list('abcde'))
```

```python
df2.loc[1, 'b'] = np.nan
```

```python
df1
```

```python
df2
```

```python
df1 + df2        #产生AN值
```

```python
df1.add(df2, fill_value = 0)      #将缺失值对象填充为0
```

```python
1 / df1
```

```python
df1.div(1)
```

```python
df1.rdiv(1)       #带r的算数方法表示翻转
```

```python
df1.reindex(columns = df2.columns, fill_value = 0)   #重建索引时，也可以指定填充值
```

```python
#DataFrame和Series间的操作
```

```python
arr = np.arange(12.).reshape((3 ,4))
```

```python
arr
```

```python
arr[0]
```

```python
arr - arr[0]           #减法在每一行都进行了操作。这就是广播机制。
```

```python
frame = pd.DataFrame(np.arange(12.).reshape((4, 3)), columns = list('bde'), index = ['Utah', 'Ohio', 'Texas', 'Oregon'])
```

```python
series = frame.iloc[0]
```

```python
frame
```

```python
series
```

```python
frame - series     #将Series的索引和DataFrame的列进行匹配，并广播到各行
```

```python
series2 = pd.Series(range(3), index = ['b', 'e', 'f'])
```

```python
series2
```

```python
frame + series2          #如果一个索引值不在DataFrame的列中，或不在Series的索引中，则对象会重建索引并形成联合
```

```python
series3 = frame['d']
```

```python
frame
```

```python
series3
```

```python
frame.sub(series3, axis = 'index')            #在行上匹配，在列上进行广播。axis匹配轴，index==0。
```
