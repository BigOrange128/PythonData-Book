```python
sdata = {'Ohio':35000, 'Texas':71000, 'Oregon':16000, 'Utah':5000}
```

```python
obj3 = Series(sdata)    #可以使用字典生成一个Series
```

```python
obj3          #索引是排序好的字典键
```

```python
states = ['California', 'Ohio', 'Oregon', 'Texas']
```

```python
obj4 = Series(sdata, index = states)       #可以将字典键按照你所想要的顺序传递给构造函数
```

```python
obj4            #因为California没有出现在sdata键中，它对应的值是NaN(not a number),这是pandas标记缺失值或NA值   Utah不在states中，被排除。
```

```python
pd.isnull(obj4)       #检查缺失数据
```

```python
obj4.notnull()
```
