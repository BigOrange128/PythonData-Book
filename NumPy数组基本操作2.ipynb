```python
#数组转置和换轴
```

```python
import numpy as np
```

```python
arr = np.arange(15).reshape((3, 5))
```

```python
arr
```

```python
arr.T        #T为数组的转置属性(适用于一二维数组)
```

```python
arr = np.random.randn(6, 3)
```

```python
arr
```

```python
np.dot(arr.T, arr)      #计算矩阵内积（矩阵乘）
```

```python
arr = np.arange(16).reshape((2, 2, 4))
```

```python
arr
```

```python
arr.transpose((1, 0, 2))
```
